```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv('./Data Collection/data.csv')
df
```

```python
valence = list(df['valence_tags'].to_numpy())
valence
def bin_val(x):
    return int(x > 5)

bin_valence = np.array(list(map(bin_val,valence)))
df.loc[:,'binary_valence'] = bin_valence
```

```python
df.to_csv('data_bin.csv')
```

## Majority Baseline for Valence

```python
count_0_maj = len([x for x in bin_valence if x == 0])
count_1_maj = len([x for x in bin_valence if x == 1])

majority_baseline = max(count_0_maj, count_1_maj) /len(bin_valence)
print("Majority Baseline is {} for majority {}".format(majority_baseline, '0' if count_0_maj > count_1_maj else '1'))
```

## Random Baseline for Valence

```python
random_baseline = [random.randint(0,1) for x in range(len(bin_valence))]

count_0_rand = len([x for x in random_baseline if x == 0])
count_1_rand = len([x for x in random_baseline if x == 1])

random_baseline = max(count_0_rand, count_1_rand) /len(bin_valence)
print("Random Baseline:",random_baseline)
```

## Average Baseline For Valence, Arousal, Dominance

```python
arousal = df['arousal_tags'].to_numpy()
dominance = df['dominance_tags'].to_numpy()

avg_valence = sum(arousal)/len(arousal)
avg_dominance = sum(dominance)/len(dominance)
avg_arousal = sum(arousal)/len(dominance)

valence_mse = 0
dominance_mse = 0
arousal_mse = 0

for values in valence:
    valence_mse += (avg_valence - values) ** 2
valence_mse = valence_mse/len(valence)
    
for values in arousal:
    arousal_mse += (avg_arousal - values) ** 2
arousal_mse = arousal_mse/len(arousal) 

for values in dominance: 
    dominance_mse += (avg_dominance - values) ** 2
dominance_mse = dominance_mse/len(dominance)

print('Average baselines: \n valence: {} \n dominance: {}\n arousal: {}\n'.format(
        valence_mse,
        dominance_mse,
        arousal_mse))
```
